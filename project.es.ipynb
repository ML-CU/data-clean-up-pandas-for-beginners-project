#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
import numpy as np
calificaciones = [8.5,9.2,7.8,8.9,10]
promedio = np.mean(calificaciones)
print(f"El promedio es {promedio}")
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
calificacion_max = max(calificaciones)
calificacion_min = min(calificaciones)

print(f'la calificación mas alta es: {calificacion_max}')
print(f'la calificación mas bajas es: {calificacion_min}')
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
ds.columns
```

```python
# TODO
casa_mas_cara = ds.loc[ds['price'].idxmax()]

direccion = casa_mas_cara['address']
precio = casa_mas_cara['price']
print(f'La casa con dirección en {direccion} es la casa mas cara y su precio es de {precio} USD')
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
casa_mas_barata = ds.loc[ds['price'].idxmin()]
direccion = casa_mas_barata['address']
precio = casa_mas_barata['price']
print(f'La casa con dirección en {direccion} es la casa mas barata y su precio es de {precio} USD')
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
# Trazar el histograma de precios
plt.figure(figsize=(15,8))
plt.hist(df_arroyomolinos['price'])
plt.xlabel('Precio')
plt.ylabel('Numero de propiedades')
plt.show()
```

1. Los precios estan concentrados entre  180000 y 400000
2. Existe pocos inmuebles, aproximadamente uno 2 que tiene un bajo precio que oscila entre 0 y 60000
3. Hay pocos inmuebles con precios muy altos, que llegan hasta casi los 600000
4. Se podria considerar como outlier los precios muy bajos y los precios muy altos.
